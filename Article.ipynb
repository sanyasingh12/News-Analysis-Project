```python
def remove_punctuation(txt):
  txt_nopunct = "".join([c for c in txt if c not in string.punctuation])
  return txt_nopunct
```

```python
import re
def tokenize(txt):
  tokens = re.split('\W+', txt)
  return tokens

data['Article_clean_tokenized'] = data['Article'].apply(tokenize)
data
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords = nltk.corpus.stopwords.words('english')
```

```python
stopwords[0:24]
```

```python
def remove_stopwords(txt_tokenized):
  txt_clean = [word for word in txt_tokenized if word not in stopwords]
  return txt_clean

data['Article_no_sw'] = data['Article_clean_tokenized'].apply(remove_stopwords)
data
```

```python
def clean_text(text):
  text = "".join([c for c in text if c not in string.punctuation])
  tokens = re.split('\W+', text)
  text = [word for word in tokens if word not in stopwords]
  return text
```

```python
data['Article_nostop'] = data['Article'].apply(clean_text)
data
```

```python
import nltk
from nltk.stem import PorterStemmer
ps = PorterStemmer()
dir(ps)
```

```python
def stemming(tokenized_text):
  text = [ps.stem(word) for word in tokenized_text]
  return text
```

```python
data['Article_stemmed'] = data['Article_nostop'].apply(stemming)
data
```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

# Load the T5 model and tokenizer
model_name = "t5-small"  # You can use other variants like t5-base or t5-large
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Define the input text
text = "Researchers discover a new treatment for cancer. This is a major breakthrough!"

# Prepare the input for the model
input_text = "summarize: " + text
input_ids = tokenizer.encode(input_text, return_tensors="pt", max_length=512, truncation=True)

# Generate the summary
summary_ids = model.generate(input_ids, max_length=50, min_length=10, length_penalty=2.0, num_beams=4, early_stopping=True)
summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print(summary)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
import pandas as pd

# Load your articles into a list
articles = [
    "Scientists develop a new solar panel that's more efficient than ever before. This could help us fight climate change!",
    "New AI technology outperforms humans in various tasks. Experts are both excited and concerned about the implications.",
    "Breakthrough in quantum computing could revolutionize various industries. The potential applications are vast.",
    # Add more articles here
]

# Load the pre-trained Sentence-BERT model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode the articles to get embeddings
embeddings = model.encode(articles)

# Use KMeans clustering to find common themes
num_clusters = 3
clustering_model = KMeans(n_clusters=num_clusters)
clustering_model.fit(embeddings)
cluster_assignments = clustering_model.labels_

# Create a DataFrame to see which article belongs to which cluster
df = pd.DataFrame({'Article': articles, 'Cluster': cluster_assignments})
print(df)

# Analyze the articles in each cluster to find common themes
for cluster in range(num_clusters):
    print(f"Cluster {cluster}:")
    print(df[df['Cluster'] == cluster]['Article'].values)
```
